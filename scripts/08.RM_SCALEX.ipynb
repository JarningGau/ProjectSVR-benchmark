## PBMC task

### Model building

```python
time0 = time.time()
adata_ref=SCALEX('../data/DISCO_hPBMC.train.counts.h5ad', 
                 batch_name = 'project_id', 
                 min_features = 0, 
                 min_cells = 3, 
                 n_top_features=2000,
                 outdir = '../ref_models/08_SCALEX/PBMC/', 
                 ignore_umap=True,
                 show = False,
                 gpu = 0)
time1 = time.time()
runtime1 = time1 - time0
```

```python
adata_ref
```

```python
sc.pp.neighbors(adata_ref, use_rep="latent")
sc.tl.umap(adata_ref)
```

```python
sc.pl.umap(
    adata_ref,
    frameon=False,
    show=False,
    color=["cell_subtype"],
    alpha=0.7
)
```

### Reference mapping

```python
time0 = time.time()
adata=SCALEX('../data/DISCO_hPBMC.test.counts.h5ad',
             batch_name='project_id',
             min_features=0, 
             min_cells=3,
             outdir='../results/08_SCALEX/PBMC_projection', 
             projection='../ref_models/08_SCALEX/PBMC/',
             show=False,
             ignore_umap=True,
             gpu=0)
time1 = time.time()
runtime2 = time1 - time0
```

```python
sc.pp.neighbors(adata, use_rep="latent")
sc.tl.umap(adata)
```

```python
adata_query = sc.read_h5ad("../data/DISCO_hPBMC.test.counts.h5ad")
```

```python
ax = sc.pl.umap(
    adata,
    frameon=False,
    show=False,
)
sc.pl.umap(
    adata[: adata_query.n_obs],
    color=["cell_subtype"],
    frameon=False,
    title="Query labels",
    ax=ax,
    alpha=0.7,
)
```

### Saving results

```python
umap_emb = pd.DataFrame(adata.obsm["X_umap"])
umap_emb.index = adata.obs.index
umap_emb.columns = ["Dim_1", "Dim_2"]

cellmeta = pd.DataFrame(adata.obs["cell_subtype"])
```

```python
ref_emb = umap_emb.loc[adata_ref.obs.index]
query_emb = umap_emb.loc[adata_query.obs.index]

ref_cellmeta = cellmeta.loc[adata_ref.obs.index]
query_cellmeta = cellmeta.loc[adata_query.obs.index]
ref_cellmeta["group"] = "reference"
query_cellmeta["group"] = "query"
ref_cellmeta.columns = ["label", "group"]
query_cellmeta.columns = ["label", "group"]
```

```python
ref_emb.to_csv("../results/08_SCALEX/PBMC_ref_embeddings.csv")
ref_cellmeta.to_csv("../results/08_SCALEX/PBMC_ref_cellmeta.csv")
query_emb.to_csv("../results/08_SCALEX/PBMC_query_embeddings.csv")
query_cellmeta.to_csv("../results/08_SCALEX/PBMC_query_cellmeta.csv")
# run time
runtime = {"model.building": [runtime1], "reference.mapping": [runtime2]}
runtime = pd.DataFrame(runtime)
runtime.to_csv("../results/08_SCALEX/PBMC_runtime.csv")
```
